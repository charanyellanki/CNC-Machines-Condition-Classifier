### Window method

```python
import numpy as np
import pandas as pd
import os
from utils import data_loader_utils # Assuming this is the module that contains datafile_read function

machines = ["M03"]
process_names = ["OP00", "OP01", "OP02", "OP03", "OP04", "OP05", "OP06", "OP07", "OP08", "OP09", "OP10", "OP11", "OP12", "OP13", "OP14"]
labels = ["good", "bad"]
path_to_dataset = "./data/"

# Define window size and stride
window_size = 10
stride = 10

# Initialize an empty list to store the results
result_df_list = []

# Iterate over each combination of machine, process_name, and label
for machine in machines:
    for process_name in process_names:
        for label in labels:
            # Construct the folder path
            folder_path = os.path.join(path_to_dataset, machine, process_name, label)

            # Check if the folder exists
            if os.path.exists(folder_path):
                # Iterate over all files in the folder
                for filename in os.listdir(folder_path):
                    # Check if the file has .h5 extension
                    if filename.endswith(".h5"):
                        # Construct the file path
                        file_path = os.path.join(folder_path, filename)

                        # Load the data from .h5 file
                        sample = data_loader_utils.datafile_read(file_path, plotting=False)

                        # Convert the data array to a DataFrame
                        df = pd.DataFrame(sample, columns=['x', 'y', 'z'])

                        # Initialize starting index for the sliding window
                        start_index = 0

                        # Initialize an empty list to store window summaries
                        window_summaries = []

                        # Iterate over the data using sliding window
                        while start_index + window_size <= len(df):
                            # Extract the window of data
                            window_data = df.iloc[start_index:start_index+window_size]

                            # Calculate the five-number summary for each column
                            summary_stats = window_data.describe().transpose()

                            # Extract the required statistics (min, 25%, 50%, 75%, max) for each column
                            summary_stats = summary_stats[['min', '25%', '50%', '75%', 'max']]

                            # Append the summary statistics to the list
                            window_summaries.append(summary_stats.values.flatten())

                            # Increment the start index by the stride
                            start_index += stride

                        # Concatenate all window summaries for the file
                        file_summary = np.concatenate(window_summaries)

                        # Flatten the summary into a single row DataFrame
                        result_row = pd.DataFrame(file_summary.reshape(1, -1))

                        # Add additional columns for machine, process_name, and label
                        result_row['machine'] = machine
                        result_row['process_name'] = process_name
                        result_row['label'] = label

                        # Append the result to the list
                        result_df_list.append(result_row)

# Concatenate all results into a single DataFrame
result_df = pd.concat(result_df_list, ignore_index=True)

print("Resultant DataFrame:")
print(result_df)


result_df.to_csv('window_2.csv', index=False)
```
